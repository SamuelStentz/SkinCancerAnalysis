```python
list_csv = ["best_validation.csv", "rfc_FF.csv", "rfc_FF.csv", "svm_FF.csv", "svm_FT.csv"]
labels = ["CNN", "RF", "SVM"]

list_csv = ["best_validation.csv", "best_validation_balanced.csv", "rfc_FF.csv",
            "rfc_FF.csv", "svm_FT.csv", "svm_FF.csv", "svm_7.csv", "best_validation_mobilenet_1.csv",
            "best_validation_mobilenet_balanced.csv"]
labels = ["CNN", "CNN-bal", "RF","RF-scaled", "SVM", "SVM-scaled", "SVM-LDA", "MobileNet", "MobileNet-bal"]
```

```python
import pandas as pd
from sklearn.metrics import *
import matplotlib.pyplot as plt

df_ls = []

for c,l in zip(list_csv, labels):
    df_ls.append(pd.read_csv("../Figures/" + c, index_col = 0))

accuracy = []
f1 = []
f1_macro = []

for df in df_ls:
    f1.append(f1_score(df["y_true"], df["y_pred"], average='weighted'))
    accuracy.append(accuracy_score(df["y_true"], df["y_pred"]))
    f1_macro.append(f1_score(df["y_true"], df["y_pred"], average='macro'))

plt.bar(labels, accuracy)
plt.ylim([.4, 1])
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.xticks(range(len(labels)), labels, rotation='vertical')
plt.savefig("../Figures/accuracies_all.png", dpi = 400)
plt.show()

plt.ylim([.4, 1])
plt.bar(labels, f1)
plt.ylabel("F1-Score")
plt.xlabel("Model")
plt.xticks(range(len(labels)), labels, rotation='vertical')
plt.savefig("../Figures/f1_all.png", dpi = 400)
plt.show()

plt.ylim([.2, 1])
plt.bar(labels, f1_macro)
plt.ylabel("F1-Score Macro")
plt.xlabel("Model")
plt.xticks(range(len(labels)), labels, rotation='vertical')
plt.savefig("../Figures/f1_macro_all.png", dpi = 400)
plt.show()
```
